# Data Visualization

# univariate Analysis

```python
sns.countplot(x='Product_Category',data=df)
plt.title("Sale of Product Category")
plt.show()
```

```python
plt.title("Year of Bike Sale")
sns.kdeplot(x="Year",data=df,shade=True)
plt.show()
```

# Bivariate Analysis

```python
sns.barplot(x='Country',y='Order_Quantity',data=df)
plt.title("Country of Order_Quantity")
plt.show()
```

```python
sns.lineplot(x="Customer_Age",y="Order_Quantity",data=df)
plt.show()
```

```python
sns.barplot(x='Customer_Gender',y='Customer_Age',data=df)
plt.show()
```

```python
sns.scatterplot(x="Cost",y="Profit",data=df)
```
